```python
import matplotlib.pyplot as plt
plt.hist(df["views"], bins=50)
plt.title("Views Distribution")
plt.show()
```

```python
df.groupby("publish_year")["views"].mean()
```

```python
df["log_views"] = np.log1p(df["views"])
df["log_likes"] = np.log1p(df["likes"])
df["log_comments"] = np.log1p(df["comments"])
```

```python
df[["log_views", "log_likes", "log_comments"]].head()
```

```python
df["likes_per_view"] = df["likes"] / df["views"]
df["comments_per_view"] = df["comments"] / df["views"]
```

```python
df["duration_min"] = df["duration_seconds"] / 60
```

```python
df["title_length"] = df["title"].apply(len)
df["word_count"] = df["title"].apply(lambda x: len(x.split()))
```

```python
df["viral"] = df["views"] > df["views"].median()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

X = df[["log_likes", "log_comments", "duration_min", "word_count", "publish_month"]]
y = df["log_views"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)
print(r2_score(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

X = df[["log_likes", "log_comments", "duration_min", "word_count"]]
y = df["viral"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf = RandomForestClassifier()
clf.fit(X_train, y_train)
pred = clf.predict(X_test)

print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

```python
import pickle

with open("model.pkl", "wb") as f:
    pickle.dump(clf, f)
```
